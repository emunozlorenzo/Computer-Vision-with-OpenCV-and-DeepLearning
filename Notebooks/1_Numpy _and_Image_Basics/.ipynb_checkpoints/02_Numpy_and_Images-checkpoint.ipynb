# Numpy and Images

- PIL is a python library to handle images

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
```

## Read Data

```python
pic = Image.open('../../data/img/00-puppy.jpg')
```

```python
type(pic)
```

## Analysing the image

```python
# Transform the image to an array
pic_arr = np.array(pic)
```

```python
type(pic_arr)
```

```python
# Shape of the array
pic_arr.shape
```

```python
# Display data as an image
plt.imshow(pic_arr);
```

## Color Channels

### Red Channel
- Red Channel Values: 0-255
- where 0 means no red (pure black) and 255 means full pure red

```python
# Red Channel
pic_red = pic_arr.copy()[:,:,0]
# Display data as an image
plt.imshow(pic_red, cmap='gray');
```

### Green Channel
- Green Channel Values: 0-255
- where 0 means no green (pure black) and 255 means full pure green

```python
# Green Channel
pic_green = pic_arr.copy()[:,:,1]
# Display data as an image
plt.imshow(pic_green, cmap='gray');
```

### Blue Channel
- Blue Channel Values: 0-255
- where 0 means no blue (pure black) and 255 means full pure blue

```python
# Blue Channel
pic_blue = pic_arr.copy()[:,:,2]
# Display data as an image
plt.imshow(pic_blue, cmap='gray');
```

## Reassign Channels

```python
pic_copy = pic_arr.copy()
```

```python
pic_copy[:,:,0] = 0 # Red channel to 0
pic_copy[:,:,2] = 0 # Blue channel to 0
plt.imshow(pic_copy);
```
